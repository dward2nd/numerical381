Computing $\pi$ using:

$$
\frac{1}{\pi} = \frac{2\sqrt{2}}{9801} \sum^\infty_{k=0} \frac{(4k)!(1103 + 26390k)}{(k!)^4 396^k}
$$

```julia
function ramanujan(n)
    sum = big(0.0)
    for k in 0:n
        sum += factorial(big(4*k)) * (1103 + 26390*k) / factorial(big(k))^4 / 396^(4*k)
    end
    sum = sum * 2 * sqrt(2) / 9801
    
    return 1 / sum
end
```

```julia
ramanujan(3)
```

```julia
x = 0:5
y = [ramanujan(k) for k in x]

plot(x, y, title="Ramanujan π value approximation", xlabel="k", ylabel="value")
```
