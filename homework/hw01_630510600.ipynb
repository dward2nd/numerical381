## ข้อที่ 1

ดูค่า $e$ จาก built-in ใน Julia โดยพิมพ์ `\euler` แล้วกด Tab นิดหน่อย หรือใช้ `MathConstants.e`

```julia
MathConstants.e
```

```julia
ℯ
```

ประกาศฟังก์ชันสำหรับประมาณค่า

```julia
function approx_e(n::Int)::BigFloat
    sum = big(0.0)
    for k in 0:n
        sum += big(1.0) / factorial(big(k))
    end
    return sum
end
```

ประมาณค่า จาก $n=1$ ถึง $n=10$ และเทียบ loss กับ built-in $e$ 

```julia
using Printf
```

```julia
for n in 1:10
    en = approx_e(n)
    @printf("approx_e(%2d) = %.16f (loss: %.16f)\n", n, en, abs(en - MathConstants.e))
end
```

Plot ความสัมพันธ์ระหว่างค่าประมาณของ $e$ และ loss

```julia
using Plots
```

```julia
x = 1:10
y = [abs(MathConstants.e - approx_e(n)) for n in x]

plot(x, y, title="Loss of ℯ within 1 ≤ n ≤ 10", xlabel="n", ylabel="loss")
```

## ข้อ 2

นิยามฟังก์ชันสำหรับการหาอนุพันธ์ของฟังก์ชันตัวแปรเดียว

```julia
function derivative(f::Function, x::Number, h::Number)::Number
    return (f(x + h) - f(x)) / h
end
```

ทดสอบด้วยการหาอนุพันธ์ของ $f(x) = x^3$ ที่ $x = 2$ ด้วย $h = 10^{-10}$

```julia
derivative(x -> x^3, 2, 10^-10)
```
